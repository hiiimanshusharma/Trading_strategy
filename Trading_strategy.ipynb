```python
tdf = rdf.groupby('Date').agg({'Open': 'first', 'Close': 'last', 'High': 'max', 'Low': 'min'}).reset_index()
tdf.to_csv('daily.csv')
tdf
```

```python
tdf['Change'] = ((tdf['Close'] - tdf['Open'])/tdf['Close'])*100
tdf
```

```python
tdf['Signal'] = ''
tdf.loc[(tdf['Change'] > 0.5) , 'Signal'] = 'buy'
tdf.loc[(tdf['Change'] < -0.5) , 'Signal'] = 'Short'
```

```python
tdf
```

```python
# Shift the 'high' and 'low' values by one row to get the values for the next row
tdf['next_high'] = tdf['High'].shift(-1)
tdf['next_low'] = tdf['Low'].shift(-1)
tdf['next_open'] = tdf['Open'].shift(-1)
tdf['next_close'] = tdf['Close'].shift(-1)
tdf['1.01 close'] = tdf['Close']*1.01
tdf['0.99 close'] = tdf['Close']*0.99
tdf
```

```python
trade_df1 = pd.DataFrame(columns=['Entry_date', 'Entry_price', 'Exit_date', 'Exit_price',  'Type', 'Profit'])

for index, row in tdf.iterrows():
  if tdf.loc[index, 'Signal'] == 'buy':
    entry_date = tdf.loc[index, 'Date']
    entry_price = tdf.loc[index, 'Close']
    exit_price = tdf.loc[index + 1, 'Close']
    exit_date = tdf.loc[index + 1, 'Date']
    profit = ((exit_price - entry_price)/exit_price)*100
    trade_df1 = pd.concat([trade_df1, pd.DataFrame({'Entry_date': entry_date, 'Type': 'long', 'Entry_price': entry_price, 'Exit_price': exit_price, 'Exit_date': exit_date, 'Profit': profit}, index=[0])], ignore_index=True)
  # elif row['Signal'] == 'Short':
  #   entry_date = tdf.loc[index, 'Date']
  #   entry_price = tdf.loc[index, 'Close']
  #   exit_price = tdf.loc[index + 1, 'Close']
  #   exit_date = tdf.loc[index + 1, 'Date']
  #   profit = ((entry_price - exit_price)/entry_price)*100
  #   trade_df1 = pd.concat([trade_df1, pd.DataFrame({'Entry_date': entry_date, 'Type': 'Short', 'Entry_price': entry_price, 'Exit_price': exit_price, 'Exit_date': exit_date, 'Profit': profit}, index=[0])], ignore_index=True)
trade_df1.to_excel('0.5%long_market.xlsx',index=False)

sum=0
for index, row in trade_df1.iterrows():
  sum += row['Profit']

print(sum)
trade_df1
```
